```python
import pandas as pd
```

```python
df = pd.read_excel(r"C:\Users\USER\Downloads\Invoices for today\Messy_Excel_Data\Messy_Excel_Data\data\data_cleaned.xlsx")
```

```python
df = df.drop_duplicates()
```

```python
df
```

```python
df["Returned"] = df["Returned"].apply(lambda x : str(x))
```

```python
df #we would like to know the number of products that were returned.
```

```python
for x in df.index:
    if df.loc[x, "Returned"] == "False":
        df.drop(x,inplace=True)
```

```python
df = df.reset_index(drop=True)
```

```python
df #288 products were returned
```

```python
df["Profit"].sum() #this amount is the total amount lost by the company because the products above where returned.
```

```python
df.plot(title="Most Returned Category", kind = 'line', x= "Category", y= "Quantity",ylabel = "Quantity",figsize=(25,10))
#looks like office supplies are being returned the most
```

```python
df.plot(title="Which Region has returned the most", kind = 'area', x= "Region", y= "Quantity",ylabel = "Quantity",figsize=(25,10))
#the western region have returned the most.
```
